# Max Salary, Average Salary, and Rating have a high Kurtosis, let's see if we can eliminate outliers

**Min Salary**

```python
outliers = df[ df['min_salary'] >= 200 ].index
df.drop(outliers, inplace=True)
df.loc[ df['min_salary'] >= 150 ]
```

```python
df.boxplot(column=['min_salary'])
```

**Max Salary**

```python
outliers = df[ df['max_salary'] >=300 ].index
df.drop(outliers, inplace=True)
df.loc[ df['max_salary'] >=250]
```

```python
df.boxplot(column=['max_salary'])
```

**Average Salary**

```python
outliers = df[df['avg_salary'] >=250].index
df.drop(outliers, inplace=True)
df.loc[ df['avg_salary'] >=200]
```

```python
outliers = df[df['avg_salary'] <=50].index
df.drop(outliers, inplace=True)
df.loc[ df['avg_salary'] <=60]
```

```python
df.boxplot(column=['avg_salary'])
```

**Rating**

```python
outliers = df[df['Rating'] < 0.1].index
df.drop(outliers, inplace=True)
df.loc[ df['Rating'] < 0.1]
```

None greater than 5:

```python
outliers = df[df['Rating'] > 5].index
df.drop(outliers, inplace=True)
df.loc[ df['Rating'] > 5]
```

```python
df.boxplot(column=['Rating'])
```

Kurtosis was successfully minimized (except company age which is dealt with later)

# Check the Percentage of Missing Data for Company Age

Check to see why company ages are less than 0:

```python
df.loc[df['company_age'] < 0]
```

**Company Age % Missing (when <0)**

```python
num_missing_Comp_Age = df['company_age']
num_missing_Comp_Age = num_missing_Comp_Age[df['company_age']<0].count()
pct_missing_Comp_Age = num_missing_Comp_Age/len(df)
pct_missing_Comp_Age
```

Replace Company Age == -1 with the median since dropping the rows would eliminate too many rows (over 10.8%).

```python
#Check average before and after
df['company_age'].describe()
```

```python
df.loc[df['company_age'] == -1, 'company_age'] = None
```

```python
df['company_age'].median()
```

```python
med = df['company_age'].median()
df['company_age'] = df['company_age'].fillna(med)
#df.head()
```

```python
df['company_age'].describe()
```

```python
num_missing_Comp_Age = df['company_age']
num_missing_Comp_Age = num_missing_Comp_Age[df['company_age']<0].count()
pct_missing_Comp_Age = num_missing_Comp_Age/len(df)
pct_missing_Comp_Age
```

**The stats for average company age are now more accurate because now the data does not include negative ages.**
