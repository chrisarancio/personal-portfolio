# Size vs Average Salary

```r
#ggplot(data, aes(x=Size, y=avg_salary)) + geom_boxplot() + theme(axis.text.x = element_text(angle = 45, vjust = 1, hjust = 1))
data %>%
  ggplot(aes(x = fct_relevel(Size, c("1 to 50 employees", "51 to 200 employees", "201 to 500 employees", "501 to 1000 employees", "1001 to 5000 employees", "5001 to 10000 employees", "10000+ employees")), y = avg_salary)) +
  geom_boxplot() +
  labs(x = "Size", y = "Average Salary")
```

```r
ggplot(data, aes(x=avg_salary)) + geom_histogram(binwidth=10) + facet_wrap(~Size)
```

```r
ggplot(data, aes(x=avg_salary)) + geom_histogram(binwidth=10)
```

```r
data %>%
  group_by(Size) %>%
  summarize(mean=mean(avg_salary), sd=sd(avg_salary), Q1=quantile(avg_salary, 0.25), Q3=quantile(avg_salary, 0.75))
```

**Although there are differences in average salaries when the company size changes, all of the average salaries are in a 10k range between 115k and 125k.**

**This is visualized in the histrograms by size category and the histogram for avarage salary. They all have a spike around the 120k mark, which is also close to the real average salary of 120.9k across all rows of data. Thus, the average salary does not appear to be heavily affected by a company's number of employees.**

# Industry vs Average Salary

```r
ggplot(data, aes(x=avg_salary)) + geom_histogram(binwidth=10) + facet_wrap(~Industry)
```

```r
ggplot(data, aes(x=Industry, y=avg_salary)) + geom_boxplot() + theme(axis.text.x = element_text(angle = 45, vjust = 1, hjust = 1))
```

**It appears that the "Architectural & Engineering Services," "Other Retail Stores," and "Electerical & Electronic Manufacturing" industries have the highest median salaries, but those industries lack a lot of data. If you observe the bar and histrogram graphs above, you will notice that those industries do not have a lot of data. This makes an assertion about the highest average salaries inconclusive. Instead, I will limit the comparisons to industries with at least 20 points of data.**

```r
industriesWithData <- data[length(unique(data$Industry)) >= 20]
industriesWithData
```

```r
industryFreq <- fct_infreq(data$Industry)
levels(industryFreq)
```

```r
industriesWithData <- data[data$Industry %in% c('BIOTECH & PHARMACEUTICALS','IT SERVICES','COMPUTER HARDWARE & SOFTWARE','AEROSPACE & DEFENSE','ENTERPRISE SOFTWARE & NETWORK SOLUTIONS','CONSULTING','STAFFING & OUTSOURCING','INSURANCE CARRIERS','UNKNOWN','ADVERTISING & MARKETING','INTERNET','HEALTH CARE SERVICES & HOSPITALS'),]
```

```r
ggplot(industriesWithData, aes(x=avg_salary)) + geom_histogram(binwidth=10) + facet_wrap(~Industry)
```

```r
ggplot(industriesWithData, aes(x=Industry, y=avg_salary)) + geom_boxplot() + theme(axis.text.x = element_text(angle = 45, vjust = 1, hjust = 1)) + stat_summary(fun="mean",color="red", shape=13)
```

**If we restrict our search to only industries with 20 or more points of data, then it appears that the Aerospace & Defense indsustry has the highest average and median of the average salaries of the industries in our dataset.**
